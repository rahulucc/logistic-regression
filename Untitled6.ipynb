```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
read=pd.read_csv(r'C:\Users\Rahul\framingham.csv')
```

```python
read.head()
```

```python
read.shape
```

```python
read=read.drop(["age"],axis=1,inplace=False)
read.head()
```

```python
read1=read.drop(read.index[[2]],axis=0)
read.head()
```

```python
read1.describe()
```

```python
read1.isnull().sum()
```

```python
read1.glucose.value_counts().index[0]
read1.glucose.mode()
```

```python
read1.iloc[:,:]=read1.iloc[:,:].apply(lambda x:x.fillna(x.value_counts().index[0]))
read1.isnull().sum
```

```python
read1.dtypes
```

```python
import statsmodels.formula.api as smf
logit_model=smf.logit('male~heartRate+diaBP+BMI+cigsPerDay+education+glucose',data=read1).fit()
```

```python
type(logit_model)
logit_model.params
```

```python
logit_model.summary()
```

```python
y_pred1=logit_model.predict(read1)
y_pred1
```

```python
read1['pred_prob']=y_pred1
read1.head()
```

```python
read1.loc[read1.pred_prob>=0.5,"PValue"]=True
read1.head()
```

```python
confusion_matrix=pd.crosstab(read1['male'],read1.PValue)
confusion_matrix
```
